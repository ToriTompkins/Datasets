```python
import networkx as nx
import pandas as pd
```

```python
relationships = pd.read_csv("relationships.csv")
display(relationships)
```

```python
graph = nx.from_pandas_edgelist(relationships, source="FromNodeId", target="ToNodeId")
```

```python
nx.draw(graph)
```

```python
nx.info(graph)
```

```python
books = pd.read_csv("books.csv")
display(books)
```

```python
attributes = books.set_index("Book ID").to_dict("index")
```

```python
nx.set_node_attributes(graph, attributes)
```

```python
[node for node, attribute in graph.nodes(data=True) if attribute['Year Written'] > 1990]  
```

```python
[attribute['Book Title'] for node, attribute in graph.nodes(data=True) if attribute['Year Written'] > 1990]  
```

```python
[node for node in graph.neighbors(1)]
```

```python
book = [node for node, attribute in graph.nodes(data=True) if attribute['Book Title'] == "The Lord of the Rings"][0]  
```

```python
[graph.nodes[node]["Book Title"] for node in graph.neighbors(book)]
```
